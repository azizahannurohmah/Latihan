**loc vs. iloc**

```python
df_train = price.sample(300)
df_train.head()
```

**loc** -> mengambil data berdasarkan nilai index, seperti pada kasus di atas adalah 484, 233, 400, 420, dan 461.

**iloc** -> mengambil data berdasarkan urutan index, seperti pada kasus di atas adalah 484 urutan index 0, 233 urutan index 1, 200 urutan index 2, dst.

```python
print(df_train.loc[484,'Parking'])
df_train.loc[484]
```

```python
print(df_train.iloc[0]['Parking'])
df_train.iloc[0]
```
